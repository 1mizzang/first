##**1. oop란**

>OOP는 객체 지향 프로그래밍(Object-Oriented Programming)의 약어입니다.

>객체 지향 프로그래밍은 소프트웨어 개발 방법론 중 하나로, 프로그램을 독립적인 객체들의 집합으로 모델링하고 객체들 간의 상호작용을 통해 프로그램을 구성하는 방식을 말합니다.<br>
OOP에서 프로그램은 객체(Object)로 구성되며, 객체는 데이터와 메서드(Method)로 구성됩니다.
<br>
이러한 객체는 클래스(Class)라는 템플릿을 기반으로 생성됩니다. 클래스는 객체를 생성하기 위한 설계도 또는 틀로, 속성과 동작을 정의하는 멤버 변수와 메서드로 구성됩니다.

>OOP의 주요 특징은 다음과 같습니다:
1. 캡슐화(Encapsulation): 관련된 데이터와 메서드를 하나의 단위로 묶어 정보 은닉을 통해 외부로부터의 접근을 제어합니다.
2. 상속(Inheritance): 클래스들 간에 상속 관계를 형성하여 코드의 재사용성을 높이고, 계층적 구조를 구성합니다.
3. 다형성(Polymorphism): 동일한 메서드를 다른 방식으로 구현하여 다양한 객체를 동일한 방법으로 다룰 수 있게 합니다.
4. 추상화(Abstraction): 복잡한 세부 사항을 감추고 필요한 기능에만 집중하여 프로그램을 구현합니다.

>OOP는 코드의 가독성과 재사용성을 높이고, 유지보수성을 향상시킬 수 있는 장점을 가지고 있어서 많은 프로그래밍 언어에서 지원하고 있습니다. 대표적인 객체 지향 프로그래밍 언어로는 Java, C++, C#, Python 등이 있습니다.

##**2. 자바의 메모리 영역에 대해서**

>자바의 메모리 영역은 JVM(Java Virtual Machine)이 실행될 때 사용되는 메모리를 구분하여 관리하는 영역입니다.<br> (JVM은 자바 프로그램의 실행을 담당하며, 다양한 메모리 영역을 사용하여 프로그램의 실행과 관련된 데이터를 저장하고 관리합니다.)

>주요한 자바 메모리 영역은 다음과 같습니다:
1. 메서드 영역(Method Area 또는 Class Area): 클래스의 구조 정보(클래스 이름, 메서드, 변수, 상수 등)를 저장하는 영역입니다. 모든 스레드가 공유하는 영역으로, JVM이 시작될 때 생성되고 프로그램이 종료될 때까지 유지됩니다.
2. 힙(Heap): 객체 인스턴스가 생성되는 영역으로, 동적으로 할당된 객체들과 배열이 저장됩니다. GC(Garbage Collector)에 의해 관리되며, 객체의 생성과 소멸이 이루어지는 곳입니다. 여러 스레드에서 공유하는 데이터를 저장하는 공간입니다.
3. 스택(Stack): 메서드 호출과 관련된 정보를 저장하는 영역입니다. 메서드 실행 시 메서드의 매개변수, 지역 변수, 임시 데이터 등이 저장되며, 메서드가 호출될 때마다 스택 프레임(Stack Frame)이 생성되고 메서드가 종료되면 해당 스택 프레임이 제거됩니다. 스택은 각 스레드마다 별도로 생성되며, 스레드의 실행 흐름에 따라 동적으로 변합니다.
4. PC 레지스터(Program Counter Register): 현재 실행 중인 명령어의 주소를 가리키는 포인터입니다. 각 스레드마다 별도로 유지되며, 스레드가 생성될 때마다 생성됩니다.
5. 네이티브 메서드 스택(Native Method Stack): 자바 프로그램에서 사용되는 네이티브 코드(다른 언어로 작성된 코드)의 정보를 저장하는 영역입니다. 자바 코드에서 호출하는 네이티브 메서드의 실행 정보를 저장합니다.

>이러한 메모리 영역들은 JVM의 구조에 따라 할당되고 관리되며, 자바 프로그램의 실행에 필요한 데이터와 실행 정보를 저장하여 효율적인 실행 환경을 제공합니다.

##**3. 왜 interface를 사용하는지**

>인터페이스는 객체 지향 프로그래밍에서 중요한 개념으로 사용됩니다. 주요한 목적은 다음과 같습니다:
1. 다중 상속의 제한: 자바는 단일 상속만을 지원하므로, 클래스가 다른 클래스로부터 상속받을 수 있는 클래스 계층 구조를 제한합니다. 하지만 인터페이스를 사용하면 클래스는 여러 개의 인터페이스를 구현할 수 있으므로, 다중 상속과 유사한 효과를 얻을 수 있습니다.
2. 클래스 간의 결합도 감소: 인터페이스를 사용하면 클래스 간의 결합도를 낮출 수 있습니다. 클래스는 인터페이스를 구현하여 원하는 기능을 제공할 수 있으며, 인터페이스에 정의된 메서드를 사용하여 클래스 간의 상호작용을 구현합니다. 이로써 클래스 간의 의존성이 줄어들고, 코드의 유연성과 재사용성이 증가합니다.
3. 다형성 구현: 인터페이스를 사용하면 다형성을 구현할 수 있습니다. 인터페이스를 구현하는 다양한 클래스는 같은 인터페이스를 통해 다양한 형태로 사용될 수 있습니다. 이는 코드의 유연성과 확장성을 높여줍니다.
4. 코드의 일관성과 가독성 향상: 인터페이스는 일관된 규칙과 명세를 제공하므로, 코드의 가독성과 이해도를 향상시킵니다. 인터페이스를 사용하여 프로그램을 설계하고 구현하면, 코드의 구조와 기능을 명확하게 표현할 수 있습니다.
5. API 설계와 관리: 인터페이스는 API(Application Programming Interface)의 일부로 사용되어 외부에 제공되는 기능과 상호작용을 정의합니다. 인터페이스를 사용하여 API를 설계하고 관리하면, 사용자들이 효과적으로 API를 활용할 수 있습니다.

>인터페이스는 코드의 유연성, 확장성, 재사용성을 높여주는 중요한 도구로 활용되며, 객체 지향 프로그래밍의 핵심 개념 중 하나입니다.



```
// 인터페이스 정의
interface Printable {
    void print();
}

// Printable 인터페이스를 구현한 클래스
class Document implements Printable {
    @Override
    public void print() {
        System.out.println("Printing document...");
    }
}

// Printable 인터페이스를 구현한 다른 클래스
class Image implements Printable {
    @Override
    public void print() {
        System.out.println("Printing image...");
    }
}

// 메인 클래스
public class Main {
    public static void main(String[] args) {
        Printable document = new Document();
        Printable image = new Image();

        document.print(); // 출력: Printing document...
        image.print();    // 출력: Printing image...
    }
}

```
>위의 예시에서 Printable 인터페이스는 print() 메소드를 선언하고 있습니다. Document 클래스와 Image 클래스는 Printable 인터페이스를 구현하고 있으므로 print() 메소드를 구현해야 합니다.
<br><br>
Main 클래스에서는 Printable 인터페이스를 구현한 객체를 생성하여 사용합니다. Printable document = new Document();와 같이 Printable 인터페이스 타입의 변수에 Document 객체를 할당할 수 있습니다. 마찬가지로 Image 객체도 Printable 인터페이스 타입의 변수에 할당할 수 있습니다.
<br><br>
인터페이스를 구현한 객체들은 인터페이스에 선언된 메소드를 구현해야 하므로 document.print()와 image.print()처럼 각 객체의 print() 메소드를 호출할 수 있습니다. 각각의 print() 메소드는 해당하는 클래스에 정의된 내용을 실행하게 됩니다.
<br><br>
인터페이스를 사용하면 다형성을 활용하여 동일한 인터페이스를 구현한 다양한 객체들을 일관된 방식으로 다룰 수 있습니다. 이를 통해 코드의 유연성과 재사용성을 높일 수 있습니다.


##**4. jdbc에 대해서 (+예시)**

>JDBC를 사용하여 데이터베이스에 접속하고 쿼리를 실행하려면 데이터베이스 벤더에서 제공하는 JDBC 드라이버가 필요합니다. 이 드라이버는 데이터베이스와의 통신을 위한 프로토콜과 기능을 제공합니다.
<br><br>
JDBC는 자바 애플리케이션과 다양한 데이터베이스 시스템 간의 상호작용을 표준화하고, 데이터베이스 연동에 대한 일관성과 이식성을 제공합니다.
<br><br>
JDBC(Java Database Connectivity)는 자바 언어를 사용하여 관계형 데이터베이스와 연동하기 위한 자바 API(응용 프로그래밍 인터페이스)입니다. JDBC는 데이터베이스에 접속하고 쿼리를 실행하며, 데이터를 관리하는 데 사용됩니다.
<br><br>
JDBC는 데이터베이스 벤더의 드라이버를 통해 데이터베이스와의 통신을 담당합니다. 각 데이터베이스 벤더는 자체 JDBC 드라이버를 제공하며, 이 드라이버는 JDBC API와 통신하는 인터페이스 역할을 합니다.


>JDBC를 사용하여 데이터베이스와 상호작용하는 과정은 일반적으로 다음 단계로 구성됩니다:
1. 드라이버 로딩: 데이터베이스 벤더가 제공하는 JDBC 드라이버를 로드합니다. 드라이버는 일반적으로 **`Class.forName()`** 메서드를 사용하여 로드됩니다.
2. 연결 설정: **`DriverManager.getConnection()`** 메서드를 사용하여 데이터베이스와의 연결을 설정합니다. 이 메서드에는 데이터베이스 URL, 사용자 이름, 비밀번호 등 연결에 필요한 정보가 전달됩니다.
3. SQL 문 실행: **`Connection`** 인터페이스를 통해 데이터베이스 연결을 얻은 후, **`Statement`** 또는 **`PreparedStatement`** 인터페이스를 사용하여 SQL 문을 실행합니다. SQL 문은 데이터베이스로 보내지고 실행 결과를 반환받을 수 있습니다.
4. 결과 처리: SQL 문 실행 결과인 **`ResultSet`**을 사용하여 데이터베이스에서 검색된 데이터를 처리합니다. **`ResultSet`**은 데이터베이스의 결과 집합을 나타내며, 데이터를 읽고 조작할 수 있는 메서드를 제공합니다.
5. 연결 종료: 데이터베이스 작업이 완료되면 **`Connection`**을 닫아 연결을 종료합니다. **`Statement`**, **`PreparedStatement`**, **`ResultSet`** 등의 자원도 명시적으로 해제해야 합니다.
<br><br>
JDBC를 사용하면 자바 애플리케이션에서 다양한 데이터베이스와 상호작용할 수 있으며, 데이터베이스에 데이터를 저장, 검색, 갱신하는 등의 작업을 수행할 수 있습니다. JDBC는 자바의 표준 데이터베이스 액세스 방법이며, 다양한 데이터베이스 시스템과의 호환성과 이식성을 제공합니다.

```java
import java.sql.*;

public class JDBExample {
    public static void main(String[] args) {
        // 데이터베이스 연결 정보
        String url = "jdbc:mysql://localhost:3306/mydatabase";
        String username = "root";
        String password = "password";

        // JDBC 연결
        try (Connection connection = DriverManager.getConnection(url, username, password)) {
            System.out.println("데이터베이스에 연결되었습니다.");

            // 쿼리 실행
            String sql = "SELECT * FROM users";
            try (Statement statement = connection.createStatement();
                 ResultSet resultSet = statement.executeQuery(sql)) {
                while (resultSet.next()) {
                    int id = resultSet.getInt("id");
                    String name = resultSet.getString("name");
                    String email = resultSet.getString("email");

                    System.out.println("ID: " + id);
                    System.out.println("이름: " + name);
                    System.out.println("이메일: " + email);
                    System.out.println("--------------------");
                }
            }
        } catch (SQLException e) {
            e.printStackTrace();
        }
    }
}
```
>위의 예시에서는 MySQL 데이터베이스에 JDBC를 사용하여 연결하고, "users" 테이블에서 데이터를 가져옵니다.<br><br>
먼저, **`DriverManager.getConnection()`** 메서드를 사용하여 데이터베이스에 연결합니다. 이때 JDBC URL, 사용자 이름 및 비밀번호를 제공해야 합니다.<br><br>
연결이 성공하면 **`Statement`** 객체를 사용하여 SQL 쿼리를 실행합니다. 위의 예시에서는 **`SELECT`** 문을 사용하여 "users" 테이블의 모든 데이터를 가져옵니다.<br><br>
**`executeQuery()`** 메서드는 **`ResultSet`** 객체를 반환하고, **`while`** 루프를 사용하여 결과 집합을 반복하여 데이터를 추출합니다. **`ResultSet`** 객체의 **`getInt()`**, **`getString()`** 등의 메서드를 사용하여 열의 값을 가져올 수 있습니다.<br><br>
이를 통해 JDBC를 사용하여 데이터베이스에 연결하고 쿼리를 실행할 수 있습니다. JDBC는 다양한 데이터베이스와 호환되며, 데이터베이스 연결 및 관리, 데이터 검색 및 조작 등 다양한 작업을 수행할 수 있는 강력한 도구입니다.

##**5. 직렬화에 대해서 설명 (+예시)**

>간략히 : 직렬화(serialization)는 객체를 일련의 바이트로 변환하는 과정을 말합니다. 이렇게 직렬화된 객체는 파일이나 네트워크를 통해 전송하거나 저장할 수 있습니다. 직렬화를 통해 객체는 그 상태를 보존하며, 나중에 필요할 때 다시 역직렬화(deserialization)하여 원래의 객체로 복원할 수 있습니다.
<br><br>자바에서는 java.io.Serializable 인터페이스를 구현함으로써 직렬화를 지원합니다. 직렬화를 위해 직렬화 가능한 클래스는 Serializable 인터페이스를 구현하고, 직렬화할 필드는 transient 키워드를 사용하여 제외시킬 수 있습니다.
<br><br>
객체를 직렬화하려면 ObjectOutputStream 클래스를 사용하여 객체를 출력하고, 역직렬화하려면 ObjectInputStream 클래스를 사용하여 객체를 입력합니다.
<br><br>
직렬화는 객체의 상태를 저장하고 공유하는데 유용합니다. 예를 들어, 객체를 파일에 저장하고 나중에 다시 불러와 사용할 수 있습니다. 또한, 분산 시스템에서 객체를 네트워크를 통해 전송하거나 다른 프로세스 간에 공유할 수도 있습니다



>직렬화(Serialization)는 객체를 바이트 스트림으로 변환하는 과정을 말합니다. 이 과정에서 객체의 상태 정보와 구조가 보존되어 저장 또는 전송할 수 있게 됩니다. 직렬화를 통해 객체는 메모리에 존재하는 상태에서 영구적인 저장이나 네트워크 전송 등의 용도로 사용될 수 있습니다.
<br><br>
직렬화는 주로 다음과 같은 상황에서 활용됩니다:
<br><br>
1. 객체의 저장: 직렬화를 통해 객체를 파일 시스템에 저장하거나 데이터베이스에 저장할 수 있습니다. 이렇게 저장된 객체는 나중에 읽어와서 사용할 수 있습니다.
2. 객체의 전송: 직렬화를 통해 객체를 네트워크를 통해 전송할 수 있습니다. 객체를 바이트 스트림으로 변환한 후, 수신 측에서는 역직렬화를 통해 원래의 객체로 복원할 수 있습니다.

>자바에서는 **`java.io.Serializable`** 인터페이스를 구현하는 클래스 객체만이 직렬화할 수 있습니다. **`Serializable`** 인터페이스는 메서드가 없는 마커 인터페이스이며, 구현 클래스가 직렬화 가능하다는 것을 나타냅니다. 직렬화 가능한 클래스는 **`ObjectOutputStream`** 클래스를 사용하여 객체를 직렬화하고, **`ObjectInputStream`** 클래스를 사용하여 객체를 역직렬화할 수 있습니다.
<br><br>
직렬화는 객체의 상태를 저장하고 전송하는 데 유용하지만, 주의할 점도 있습니다. 직렬화된 객체는 버전 관리, 보안 등의 측면에서 주의가 필요합니다. 직렬화된 객체를 역직렬화할 때는 원본 객체와 버전이 호환되는지 확인해야 하며, 신뢰할 수 없는 데이터를 역직렬화하는 경우 보안 취약점이 될 수 있습니다.

```java
import java.io.*;

public class SerializationExample {
    public static void main(String[] args) {
        // 직렬화할 객체 생성
        Person person = new Person("John Doe", 25);

        // 객체를 파일에 저장 (직렬화)
        try (FileOutputStream fileOut = new FileOutputStream("person.ser");
             ObjectOutputStream objectOut = new ObjectOutputStream(fileOut)) {
            objectOut.writeObject(person);
            System.out.println("직렬화: person 객체가 파일에 저장되었습니다.");
        } catch (IOException e) {
            e.printStackTrace();
        }

        // 파일에서 객체를 읽어옴 (역직렬화)
        try (FileInputStream fileIn = new FileInputStream("person.ser");
             ObjectInputStream objectIn = new ObjectInputStream(fileIn)) {
            Person serializedPerson = (Person) objectIn.readObject();
            System.out.println("역직렬화: 파일에서 person 객체를 읽어왔습니다.");
            System.out.println("이름: " + serializedPerson.getName());
            System.out.println("나이: " + serializedPerson.getAge());
        } catch (IOException | ClassNotFoundException e) {
            e.printStackTrace();
        }
    }
}

class Person implements Serializable {
    private String name;
    private int age;

    public Person(String name, int age) {
        this.name = name;
        this.age = age;
    }

    public String getName() {
        return name;
    }

    public int getAge() {
        return age;
    }
}
```

>위의 예시에서는 **`Person`** 클래스를 직렬화하고 역직렬화합니다. **`Person`** 클래스는 **`Serializable`** 인터페이스를 구현하여 직렬화가 가능하도록 표시되어 있습니다.
<br><br>
**`Person`** 객체를 먼저 직렬화하여 파일에 저장하고, 그 다음에 파일에서 객체를 읽어와서 역직렬화합니다. 역직렬화된 객체를 사용하여 객체의 데이터에 접근할 수 있습니다.
<br><br>
위의 예시를 실행하면 **`person.ser`**라는 파일에 **`Person`** 객체가 저장되고, 그 후에 역직렬화하여 객체의 정보를 출력합니다. 이를 통해 객체의 상태를 유지하면서 파일이나 네트워크를 통해 데이터를 전송하거나 저장할 수 있습니다.



##**6. serialVersionUID를 선언해야하는 이유 (+예시)**

>**`serialVersionUID`**는 직렬화된 객체의 버전을 식별하기 위한 고유한 식별자입니다. 이를 선언하는 이유는 다음과 같습니다:
<br><br>
1. 버전 관리: **`serialVersionUID`**를 명시적으로 선언함으로써, 직렬화된 객체의 버전을 명시적으로 관리할 수 있습니다. 객체의 구조나 속성이 변경되었을 때, 버전이 다른 객체 간에 역직렬화를 시도하면 호환성 문제가 발생할 수 있습니다. **`serialVersionUID`**를 사용하여 버전이 일치하는지 확인하고, 일치하지 않으면 호환되지 않는 버전임을 알려줄 수 있습니다.
2. 역직렬화의 안정성: **`serialVersionUID`**를 명시적으로 선언하면, 직렬화된 객체의 클래스 구조가 변경되어도 역직렬화 과정에서 예외가 발생하지 않습니다. 역직렬화 시에 자바는 클래스 구조와 **`serialVersionUID`** 값을 비교하여 호환성을 검사하고, 일치하지 않으면 **`InvalidClassException`** 예외를 발생시킵니다. 이를 통해 역직렬화 과정에서의 예기치 않은 문제를 방지할 수 있습니다.
3. 보안: **`serialVersionUID`**를 명시적으로 선언하여 객체의 버전을 식별함으로써, 악의적인 변경이나 조작을 감지할 수 있습니다. 만약 직렬화된 객체의 **`serialVersionUID`**가 변경되었다면, 객체의 무결성에 문제가 있을 수 있으며, 이는 데이터 무결성을 보장하는 데 도움을 줍니다.
<br><br>
따라서 **`serialVersionUID`**를 선언함으로써 버전 관리, 역직렬화의 안정성, 보안 측면에서 이점을 얻을 수 있습니다. 주의할 점은 **`serialVersionUID`**가 변경되면 호환성이 깨지므로, 변경되지 않도록 주의해야 합니다.
아래는 `serialVersionUID`를 선언한 예시입니다:

```java
import java.io.Serializable;

public class MyClass implements Serializable {
    private static final long serialVersionUID = 123456789L;
    // 클래스의 멤버 변수와 메서드들...
}

```

>위 예시에서 **`MyClass`**는 직렬화가 가능한(serializable) 클래스입니다. **`serialVersionUID`**는 **`123456789L`**로 명시적으로 선언되어 있습니다. 이를 통해 객체의 직렬화 버전을 식별할 수 있습니다.
<br><br>
만약 **`MyClass`**의 구조나 멤버 변수가 변경되었을 때, **`serialVersionUID`**를 변경하지 않는다면 호환성 문제가 발생할 수 있습니다. 이를 방지하기 위해 클래스를 업데이트한 후 **`serialVersionUID`**를 새로운 값으로 업데이트해야 합니다. 이렇게 하면 이전 버전과 새 버전 간의 호환성이 유지됩니다.
<br><br>
주의할 점은 **`serialVersionUID`** 값은 **`static final long`** 타입이어야 하며, 반드시 **`private`** 접근 제어자를 가져야 합니다. 또한, **`serialVersionUID`** 값을 일련번호처럼 임의로 지정할 수 있지만, 충돌을 방지하기 위해 고유한 값으로 사용해야 합니다.

##**7. 리플렉션이란 (+예시)**

>리플렉션(Reflection)은 프로그램의 실행 중에 자기 자신의 구조를 분석하고 수정할 수 있는 능력을 말합니다. 즉, 리플렉션은 실행 중인 클래스의 정보를 동적으로 가져오고, 클래스의 멤버 변수, 메서드, 생성자 등에 접근하고 조작할 수 있는 기능을 제공합니다.

>자바에서 리플렉션은 **`java.lang.reflect`** 패키지에 있는 클래스와 인터페이스를 통해 구현됩니다. 주요 클래스로는 **`Class`**, **`Field`**, **`Method`**, **`Constructor`** 등이 있습니다. 이러한 클래스를 사용하여 실행 중인 클래스의 정보를 가져올 수 있으며, 필드의 값을 읽고 수정하거나, 메서드를 호출하거나 생성자를 호출하는 등의 작업을 수행할 수 있습니다.
<br><br>
리플렉션은 다음과 같은 상황에서 유용하게 활용될 수 있습니다:
<br><br>
- 런타임에 동적으로 클래스를 로드하고 인스턴스를 생성할 때
- 클래스의 구조를 분석하여 메서드나 필드의 이름, 타입, 접근 제어자 등을 가져올 때
- 메서드를 동적으로 호출하고 실행할 때
- 주어진 객체의 타입을 동적으로 확인하거나, 객체를 타입별로 다룰 때

리플렉션은 강력한 기능이지만 성능상의 이슈와 보안상의 문제가 있을 수 있으므로, 신중하게 사용해야 합니다. 일반적으로 리플렉션은 특수한 경우에만 사용되며, 일반적인 프로그래밍 작업에는 필요하지 않을 수 있습니다.

```java
import java.lang.reflect.Method;

public class ReflectionExample {
    public static void main(String[] args) throws Exception {
        // 클래스 이름으로 Class 객체 가져오기
        Class<?> myClass = Class.forName("com.example.MyClass");

        // 클래스의 메서드 가져오기
        Method[] methods = myClass.getDeclaredMethods();

        // 메서드 실행
        for (Method method : methods) {
            if (method.getName().equals("myMethod")) {
                // 메서드 호출을 위해 인스턴스 생성
                Object instance = myClass.getDeclaredConstructor().newInstance();
                // 메서드 호출
                method.invoke(instance, "Hello, Reflection!");
                break;
            }
        }
    }
}

class MyClass {
    public void myMethod(String message) {
        System.out.println("MyMethod: " + message);
    }
}
```

>위의 예시에서는 **`Class.forName()`** 메서드를 사용하여 동적으로 클래스 **`MyClass`**의 **`Class`** 객체를 가져옵니다. 그 후, **`getDeclaredMethods()`** 메서드를 사용하여 클래스의 모든 메서드를 가져옵니다. 메서드 중에서 이름이 "myMethod"인 메서드를 찾고, 해당 메서드를 실행합니다.
<br><br>
이 예시에서는 **`myMethod`** 메서드에 문자열을 전달하여 출력하도록 하였습니다. 이렇게 리플렉션을 사용하면 런타임에 클래스의 메서드를 동적으로 조작할 수 있습니다.

##**8. 자바의 클래스 멤버 변수 초기화 순서에 대해서**

>자바에서 클래스의 멤버 변수 초기화는 다음 순서에 따라 이루어집니다:
<br><br>
멤버 변수 선언: 클래스 내부에 멤버 변수를 선언합니다. 이 단계에서 멤버 변수는 해당 데이터 타입의 기본값으로 초기화됩니다. 예를 들어, int 타입의 멤버 변수는 0으로, boolean 타입의 멤버 변수는 false로 초기화됩니다.
<br><br>
명시적 초기화: 멤버 변수를 선언할 때 초기값을 명시적으로 지정할 수 있습니다. 이 경우 명시한 값으로 멤버 변수가 초기화됩니다. 예를 들어, int count = 10;과 같이 선언하면 count 멤버 변수는 10으로 초기화됩니다.
<br><br>
인스턴스 초기화 블록: 인스턴스 초기화 블록은 중괄호 {}로 둘러싸인 코드 블록으로, 인스턴스가 생성될 때 실행됩니다. 인스턴스 초기화 블록은 new 키워드로 객체를 생성할 때마다 실행됩니다. 멤버 변수의 초기화에 복잡한 로직이 필요한 경우에 사용됩니다.
<br><br>
생성자: 생성자는 객체를 초기화하는 데 사용됩니다. 생성자는 new 키워드로 객체를 생성할 때 호출되며, 멤버 변수의 초기화에 사용될 수 있습니다. 생성자는 인자를 받아서 멤버 변수에 값을 할당하거나 다른 초기화 작업을 수행할 수 있습니다.
<br><br>
클래스의 멤버 변수 초기화는 위의 순서에 따라 이루어지며, 초기화 과정에서 멤버 변수들은 각각의 초기값 또는 명시한 값으로 초기화됩니다. 이후에 인스턴스 초기화 블록과 생성자에서 추가적인 초기화 작업이 수행될 수 있습니다.
물건을 나타내는 **`Product`** 클래스의 멤버 변수 초기화 예시를 보여드리겠습니다:

```java
public class Product {
    // 멤버 변수 선언 (기본값으로 초기화)
    private int id;         // int 타입의 기본값 0
    private String name;    // String 타입의 기본값 null
    private double price;   // double 타입의 기본값 0.0
    private boolean active; // boolean 타입의 기본값 false
    
    // 명시적 초기화
    private int quantity = 1; // 초기값으로 1
    
    // 인스턴스 초기화 블록
    {
        // 복잡한 로직을 통한 초기화
        if (name == null) {
            name = "Unknown";
        }
    }
    
    // 생성자
    public Product(int id, String name, double price, boolean active) {
        this.id = id;
        this.name = name;
        this.price = price;
        this.active = active;
    }
    
    // Getter와 Setter 메서드는 생략
    
    // toString 메서드
    @Override
    public String toString() {
        return "Product [id=" + id + ", name=" + name + ", price=" + price + ", active=" + active + ", quantity=" + quantity + "]";
    }
}
```

>위의 예시에서 **`Product`** 클래스의 멤버 변수는 초기값으로 기본값을 가지고 있습니다. **`quantity`** 변수는 명시적으로 1로 초기화되어 있습니다. 인스턴스 초기화 블록은 **`name`** 변수가 **`null`**인 경우 "Unknown"으로 초기화하는 복잡한 로직을 포함하고 있습니다. 생성자를 통해 멤버 변수에 값을 할당할 수 있습니다. **`toString`** 메서드는 객체의 정보를 문자열로 반환하는데, 해당 예시에서는 멤버 변수들의 값을 포함하여 반환합니다.





##**9. servlet과 jsp의 공통점과 차이점**

Servlet과 JSP는 모두 Java를 기반으로 하는 웹 애플리케이션 개발을 위한 기술입니다. 아래는 Servlet과 JSP의 공통점과 차이점에 대한 설명입니다:

공통점:

웹 애플리케이션 개발을 위해 Java 언어를 사용합니다.
서버 측에서 실행되며, 동적인 웹 페이지를 생성하고 제공합니다.
웹 애플리케이션의 비즈니스 로직을 구현하고 데이터 처리를 할 수 있습니다.
클라이언트 요청에 대한 응답을 생성하여 전송할 수 있습니다.
HTTP 프로토콜을 기반으로 동작하며, 웹 요청과 응답을 처리합니다.
차이점:

구조: Servlet은 Java 클래스로 이루어져 있으며, Java 코드 내에서 HTML 코드를 작성해야 합니다. JSP는 HTML 문서 내에 Java 코드를 삽입하여 동적인 처리를 할 수 있도록 합니다.
문법: Servlet은 Java 코드를 사용하기 때문에 Java의 문법과 구조를 따릅니다. JSP는 HTML 문서 내에 Java 코드를 삽입하기 위해 특수한 태그(<% %>)를 사용합니다.
디자인과 로직의 분리: Servlet은 Java 코드와 HTML 코드가 섞여 있기 때문에 디자인과 로직이 분리되지 않을 수 있습니다. JSP는 HTML과 Java 코드를 분리하여 작성할 수 있으므로, 디자이너와 개발자가 병렬로 작업할 수 있습니다.
유지보수와 가독성: JSP는 HTML 문서 내에 Java 코드를 포함하기 때문에 디자이너가 HTML 부분을 수정할 수 있습니다. Servlet은 Java 코드를 작성하는 것이므로 개발자에 의해 수정되어야 합니다. 따라서 JSP는 유지보수와 가독성이 더 용이합니다.
재사용성: Servlet은 Java 클래스로 이루어져 있어 다른 웹 애플리케이션에서도 재사용할 수 있습니다. JSP는 HTML과 Java 코드가 혼합되어 있기 때문에 다른 프로젝트에 그대로 적용하기는 어렵습니다.
Servlet과 JSP는 각각의 특징과 장단점을 가지고 있으며, 프로젝트의 요구사항과 개발자의 선호도에 따라 적절한 기술을 선택하여 사용할 수 있습니다.

```java
servlet의 예시
import java.io.IOException;
import javax.servlet.ServletException;
import javax.servlet.http.HttpServlet;
import javax.servlet.http.HttpServletRequest;
import javax.servlet.http.HttpServletResponse;

public class HelloServlet extends HttpServlet {
    protected void doGet(HttpServletRequest request, HttpServletResponse response) throws ServletException, IOException {
        response.setContentType("text/html");
        response.getWriter().println("<h1>Hello, Servlet!</h1>");
    }
}
```

```java
jsp의예시
<!-- hello.jsp -->
<!DOCTYPE html>
<html>
<head>
    <title>Hello JSP</title>
</head>
<body>
    <h1>Hello, JSP!</h1>
    <% String message = "Hello from JSP!"; %>
    <p><%= message %></p>
</body>
</html>
```

위의 예시에서 Servlet은 Java 클래스로 작성되었으며, **`doGet()`** 메소드를 오버라이딩하여 클라이언트의 GET 요청에 대한 처리를 합니다. 응답으로 "Hello, Servlet!"이라는 제목이 포함된 HTML 문서를 생성하여 클라이언트에게 전송합니다.

JSP는 HTML 문서 내에 **`<% %>`** 태그를 사용하여 Java 코드를 삽입합니다. **`hello.jsp`**는 간단한 HTML 문서이며, JSP 코드를 이용하여 "Hello, JSP!"라는 제목과 "Hello from JSP!"라는 메시지를 출력합니다. JSP 코드는 **`<%= %>`** 태그를 사용하여 변수 **`message`**의 값을 출력합니다.

이는 간단한 예시이며, 실제로는 더 복잡한 비즈니스 로직과 데이터 처리를 포함하는 웹 애플리케이션을 개발할 수 있습니다.

##**10. 제네릭에 대해서**(+예시)

>제네릭(Generic)은 자바에서 컬렉션 클래스나 메소드에서 사용되는 타입 매개변수화 기법입니다. 제네릭을 사용하면 타입 안정성을 보장하고 코드의 재사용성과 가독성을 향상시킬 수 있습니다.

>제네릭을 사용하면 클래스나 메소드를 선언할 때 타입 매개변수를 사용하여 클래스나 메소드 내에서 사용되는 데이터 타입을 일반화할 수 있습니다. 이렇게 정의된 타입 매개변수는 실제 사용될 때 구체적인 타입으로 대체됩니다.
<br><br>
제네릭을 사용하면 다음과 같은 이점을 얻을 수 있습니다:
<br><br>
타입 안정성: 컴파일 시에 타입 체크를 수행하여 잘못된 타입 사용으로 인한 런타임 오류를 방지합니다.
재사용성: 동일한 로직을 다양한 타입에 대해 재사용할 수 있습니다.
가독성: 제네릭을 사용하면 코드의 의도가 명확하게 드러나므로 가독성이 향상됩니다.
제네릭을 사용하는 예시로는 컬렉션 클래스인 ArrayList를 들 수 있습니다. ArrayList는 요소들을 저장하기 위한 동적 배열 구조를 제공하는데, 제네릭을 사용하여 저장할 요소의 타입을 지정할 수 있습니다. 이를 통해 ArrayList에 저장되는 요소의 타입이 명확해지고 타입 체크를 할 수 있습니다.

```java
예시
ArrayList<String> strings = new ArrayList<String>();
strings.add("Hello");
strings.add("World");
String str = strings.get(0);
System.out.println(str); // 출력: Hello
```

>위의 예시에서 ArrayList를 선언할 때 <String>을 사용하여 타입 매개변수로 String을 지정했습니다. 따라서 ArrayList에는 String 타입의 요소만 저장할 수 있습니다. strings.add("Hello")와 같이 문자열을 추가하고, strings.get(0)를 통해 첫 번째 요소를 가져올 수 있습니다. 이때 반환되는 값은 String 타입이므로 변수 str에 할당하여 출력할 수 있습니다.






##**11. 컬렉션클래스에서 제네릭을 사용하는 이유는 무엇일까?**(+예시)

>타입 안정성(Type Safety): 제네릭을 사용하면 컬렉션에 저장되는 요소의 타입을 컴파일 시점에서 체크할 수 있습니다. 컴파일러는 요소의 타입을 확인하고 컬렉션에 잘못된 타입의 요소가 추가되는 것을 방지합니다. 이를 통해 실행 시점에 발생할 수 있는 타입 관련 오류를 미리 방지할 수 있습니다.

코드 가독성과 유지 보수성: 제네릭을 사용하면 코드가 더욱 명확하고 읽기 쉬워집니다. 요소의 타입을 명시적으로 선언하므로 다른 개발자가 코드를 이해하고 유지 보수하는 데 도움이 됩니다.

타입 변환의 번거로움 감소: 제네릭을 사용하면 컬렉션에서 요소를 꺼내거나 추가할 때 타입 변환을 수동으로 처리할 필요가 없습니다. 컬렉션에 저장된 요소의 타입이 이미 제네릭으로 지정되어 있기 때문에 자동으로 타입 변환이 이루어집니다.

컴파일 시간 오류 검출: 제네릭을 사용하면 컴파일 시점에서 오류를 검출할 수 있습니다. 잘못된 타입의 요소를 컬렉션에 추가하려고 하거나, 잘못된 타입의 요소를 꺼내려고 할 때 컴파일러가 경고나 오류를 표시합니다. 이를 통해 런타임 오류를 사전에 방지할 수 있습니다.

코드 재사용성: 제네릭을 사용하면 여러 종류의 객체를 다루는 일반적인 컬렉션 클래스를 작성할 수 있습니다. 타입 매개변수를 사용하여 컬렉션 클래스를 유연하게 구성할 수 있으며, 다양한 타입에 대해 재사용할 수 있는 범용적인 코드를 작성할 수 있습니다.

이러한 이유들로 인해 제네릭은 컬렉션 클래스에서 많이 사용되며, 타입 안정성과 코드 품질을 향상시키는 데 중요한 역할을 합니다.


```
import java.util.ArrayList;

public class Main {
    public static void main(String[] args) {
        // String 타입 요소를 저장하는 ArrayList 생성
        ArrayList<String> names = new ArrayList<>();

        // 요소 추가
        names.add("Alice");
        names.add("Bob");
        names.add("Charlie");

        // 요소 접근
        String firstPerson = names.get(0);
        System.out.println("First person: " + firstPerson);

        // 반복문을 통한 요소 출력
        for (String name : names) {
            System.out.println(name);
        }
    }
}
```


위의 예시에서 ArrayList는 제네릭을 사용하여 String 타입의 요소만을 저장할 수 있도록 선언되었습니다. 이로 인해 ArrayList에 잘못된 타입의 요소를 추가하려고 하면 컴파일 오류가 발생합니다. 또한, names.get(0)를 통해 요소를 꺼낼 때도 별도의 타입 변환 과정이 필요하지 않습니다.

제네릭을 사용함으로써 코드의 가독성과 안정성이 향상되었으며, 컴파일러가 타입 체크를 수행하여 오류를 사전에 방지할 수 있게 되었습니다.

##**12. pojo란 무엇인가**(+예시)

POJO(Plain Old Java Object)는 자바에서 일반적으로 사용되는 객체를 나타내는 용어입니다. POJO는 다음과 같은 특징을 갖습니다:

간단하고 가벼운 객체: POJO는 특정 프레임워크나 라이브러리에 종속되지 않는 일반적인 자바 객체입니다. 복잡한 상속 구조나 특정 인터페이스의 구현을 필요로 하지 않으며, 필요한 필드와 메서드를 갖는 단순한 구조를 가지고 있습니다.

Getter 및 Setter 메서드: POJO는 필드에 접근하기 위해 Getter와 Setter 메서드를 제공합니다. 이를 통해 데이터를 설정하고 가져올 수 있으며, 일반적으로 자바 빈(Java Bean) 규약을 따릅니다.

직렬화 가능: POJO는 직렬화(Serialization)를 지원하여 객체의 상태를 저장하고 전송할 수 있습니다. 직렬화를 위해 java.io.Serializable 인터페이스를 구현할 수 있습니다.

프레임워크와 독립적: POJO는 특정 프레임워크나 라이브러리에 종속되지 않습니다. 이는 POJO를 다른 환경에서도 재사용할 수 있고, 테스트와 유지보수가 용이하다는 장점을 가지게 합니다.

POJO는 객체지향적인 설계 원칙을 따르며, 단순하고 유연한 구조를 가지고 있습니다. 일반적으로 데이터 전달이나 저장을 위해 사용되며, 자바의 다양한 영역에서 활용될 수 있습니다.




```
public class Person {
    private String name;
    private int age;

    public Person() {
        // 기본 생성자
    }

    public Person(String name, int age) {
        this.name = name;
        this.age = age;
    }

    // Getter 메서드
    public String getName() {
        return name;
    }

    // Setter 메서드
    public void setName(String name) {
        this.name = name;
    }

    public int getAge() {
        return age;
    }

    public void setAge(int age) {
        this.age = age;
    }
}

```
위의 Person 클래스는 POJO의 특징을 따르고 있습니다. 필드인 name과 age는 프라이빗 접근 제어자를 가지고 있으며, Getter와 Setter 메서드를 통해 접근할 수 있습니다.

POJO 클래스는 다른 클래스에서도 독립적으로 사용될 수 있으며, 다양한 용도로 활용될 수 있습니다. 예를 들어, 데이터 전달을 위한 DTO(Data Transfer Object)로 사용하거나, 데이터베이스의 레코드와 매핑되는 엔티티 클래스로 사용할 수 있습니다. POJO 클래스는 간단하고 가벼우며, 프레임워크나 라이브러리에 종속되지 않는 일반적인 자바 객체로서 유지보수와 테스트가 용이합니다.


##**13. 박싱과 언박싱에 대해 설명하시오**(+예시)

박싱(Boxing)과 언박싱(Unboxing)은 자바에서 기본 데이터 타입과 그에 대응하는 참조 타입(Wrapper 클래스) 간의 변환을 말합니다. 아래에서 박싱과 언박싱에 대해 설명하고 공통점과 차이점을 알려드리겠습니다.

박싱(Boxing)

박싱은 기본 데이터 타입을 해당하는 참조 타입으로 변환하는 과정입니다. 기본 데이터 타입의 값을 해당하는 Wrapper 클래스의 인스턴스로 감싸는 것을 의미합니다. 이로써 기본 데이터 타입의 값이 객체로 다뤄질 수 있게 됩니다.

예를 들어, int 타입의 값을 Integer 클래스의 객체로 박싱하는 것은 다음과 같습니다:



```
int number = 10;
Integer boxedNumber = Integer.valueOf(number);

```
언박싱(Unboxing)

언박싱은 박싱된 참조 타입을 해당하는 기본 데이터 타입으로 변환하는 과정입니다. 참조 타입의 값을 해당하는 기본 데이터 타입으로 추출하는 것을 의미합니다.

예를 들어, Integer 객체에서 int 타입의 값으로 언박싱하는 것은 다음과 같습니다:


```
Integer boxedNumber = Integer.valueOf(20);
int number = boxedNumber.intValue();

```
공통점

박싱과 언박싱은 기본 데이터 타입과 참조 타입 간의 변환을 수행합니다.
박싱된 값은 참조 타입으로 다뤄지기 때문에 객체 지향적인 기능을 활용할 수 있습니다.
언박싱된 값은 기본 데이터 타입으로 다뤄질 수 있어 일반적인 연산 및 처리에 용이합니다.
차이점

박싱은 기본 데이터 타입의 값을 참조 타입으로 변환하는 것이며, 언박싱은 참조 타입의 값을 기본 데이터 타입으로 변환하는 것입니다.
박싱은 valueOf() 메서드를 사용하여 수행되고, 언박싱은 xxxValue() 메서드를 사용하여 수행됩니다. (여기서 xxx는 해당하는 기본 데이터 타입을 나타냅니다.)
박싱과 언박싱은 자동으로 처리되는 오토박싱(Autoboxing)과 오토언박싱(Auto-unboxing) 기능을 통해 더욱 간편하게 사용할 수 있습니다. 이를 통해 필요한 경우에 자동으로 박싱 및 언박싱이 수행되어 코드를 간결하게 작성할 수 있습니다.






##**14. 데드락에 대한 설명과 해결방안을 얘기해주세요**(+예시)

데드락(Deadlock)은 멀티스레드 환경에서 발생할 수 있는 상황으로, 두 개 이상의 스레드가 서로가 점유한 자원을 기다리면서 무한히 대기하는 상태를 말합니다. 이 상태에서는 어떠한 스레드도 진행하지 못하고 프로그램이 멈추게 됩니다.

데드락이 발생하려면 다음의 네 가지 조건이 동시에 충족되어야 합니다:

상호 배제(Mutual Exclusion): 자원은 동시에 하나의 스레드만이 점유할 수 있어야 합니다. 다른 스레드는 접근할 수 없습니다.
점유와 대기(Hold and Wait): 스레드가 최소한 하나의 자원을 점유한 상태에서 다른 스레드가 점유한 자원을 기다리고 있어야 합니다.
비선점(No Preemption): 스레드가 다른 스레드가 점유한 자원을 강제로 빼앗을 수 없어야 합니다. 자원을 사용을 마치기 전까지는 다른 스레드가 점유할 수 없습니다.
순환 대기(Circular Wait): 두 개 이상의 스레드가 자원을 점유하기 위해 서로를 기다리고 있어야 합니다. 예를 들어, 스레드 A는 스레드 B가 점유한 자원을 기다리고, 동시에 스레드 B는 스레드 A가 점유한 자원을 기다리는 상태입니다.
데드락은 시스템의 자원을 효율적으로 사용하지 못하고, 프로그램의 진행을 멈추게 만들어 성능 저하나 비정상적인 동작을 초래할 수 있습니다. 따라서 데드락을 예방하고 해결하기 위한 적절한 방법과 알고리즘을 사용하여 문제를 해결해야 합니다.


데드락의 예시로는 "교착 상태의 식사하는 철학자들" 문제가 자주 이용됩니다. 이 문제는 다섯 명의 철학자가 원형으로 앉아 있고, 각자 앞에는 젓가락이 있습니다. 철학자들은 생각을 하다가 배가 고프면 식사를 합니다. 그런데 철학자들은 오른쪽과 왼쪽에 있는 젓가락을 동시에 집을 수 없어야 합니다.



```
// 철학자 클래스
class Philosopher {
    private Object leftChopstick;
    private Object rightChopstick;

    public Philosopher(Object leftChopstick, Object rightChopstick) {
        this.leftChopstick = leftChopstick;
        this.rightChopstick = rightChopstick;
    }

    public void eat() {
        synchronized (leftChopstick) {
            // 왼쪽 젓가락을 집는다
            synchronized (rightChopstick) {
                // 오른쪽 젓가락을 집는다
                // 식사를 한다
            }
            // 오른쪽 젓가락을 놓는다
        }
        // 왼쪽 젓가락을 놓는다
    }

    public void think() {
        // 생각을 한다
    }
}

// 철학자 객체들과 젓가락 객체들 생성
Philosopher philosopher1 = new Philosopher(chopstick1, chopstick2);
Philosopher philosopher2 = new Philosopher(chopstick2, chopstick3);
Philosopher philosopher3 = new Philosopher(chopstick3, chopstick4);
Philosopher philosopher4 = new Philosopher(chopstick4, chopstick5);
Philosopher philosopher5 = new Philosopher(chopstick5, chopstick1);

// 철학자들이 식사를 하고 생각하는 과정
while (true) {
    philosopher1.eat();
    philosopher1.think();

    philosopher2.eat();
    philosopher2.think();

    philosopher3.eat();
    philosopher3.think();

    philosopher4.eat();
    philosopher4.think();

    philosopher5.eat();
    philosopher5.think();
}

```

위의 코드에서 각 철학자는 자신의 왼쪽과 오른쪽에 있는 젓가락을 사용해 식사를 합니다. 그런데 모든 철학자가 동시에 자신의 왼쪽 젓가락을 잡으면서 오른쪽 젓가락을 기다리는 상황이 발생할 수 있습니다. 이런 상황에서 철학자들은 서로가 자신의 오른쪽 젓가락을 놓지 않는 한 계속해서 대기 상태에 빠지게 되어 데드락이 발생합니다. 이 문제를 해결하기 위해서는 젓가락을 동시에 잡을 수 있는 조건을 만들어주거나, 철학자들이 특정한 순서로 젓가락을 잡도록 제어해야 합니다.



##**15. 상속과 합성의 차이점과 공통점에 대해 설명**

>상속과 합성은 객체 지향 프로그래밍에서 클래스 간의 관계를 표현하는 데 사용되는 개념입니다. 이 둘은 다음과 같은 차이점과 공통점이 있습니다.

>차이점:
<br>
관계 유형: 상속은 "is-a" 관계를 나타내는데 사용되고, 합성은 "has-a" 관계를 나타내는데 사용됩니다. 상속은 클래스 간의 계층 구조를 형성하며, 자식 클래스가 부모 클래스의 특성과 동작을 상속받는 것을 의미합니다. 반면에 합성은 클래스가 다른 클래스를 구성 요소로 가지는 것을 의미하며, 객체 간의 협력 관계를 표현합니다.
<br><br>
확장성: 상속은 계층적인 확장 구조를 가지기 때문에 새로운 클래스를 추가하거나 기존 클래스를 수정하여 기능을 확장할 수 있습니다. 합성은 객체 간의 협력을 통해 기능을 확장하므로, 객체의 구성 요소를 변경하거나 새로운 객체를 추가함으로써 확장될 수 있습니다.
<br><br>
클래스 간의 결합도: 상속은 부모 클래스와 자식 클래스 사이의 강한 결합도를 가지고 있습니다. 자식 클래스는 부모 클래스에 의존하며, 부모 클래스의 변경이 자식 클래스에 영향을 미칠 수 있습니다. 합성은 객체 간의 느슨한 결합도를 가지고 있습니다. 객체는 다른 객체를 포함하고 있지만, 포함된 객체의 변경이 포함하는 객체에 직접적인 영향을 주지는 않습니다.
<br><br>

>공통점:
<br>
재사용성: 상속과 합성은 코드의 재사용을 가능하게 합니다. 상속은 부모 클래스의 특성과 동작을 자식 클래스가 상속받아 사용할 수 있으며, 합성은 다른 클래스를 구성 요소로 사용함으로써 객체의 재사용을 촉진합니다.
<br><br>
다형성: 상속과 합성은 다형성을 지원합니다. 다형성은 동일한 타입으로 다양한 구현을 사용할 수 있는 능력을 의미하며, 상속과 합성을 통해 다형성을 구현할 수 있습니다.

>결론적으로, 상속과 합성은 객체 간의 관계를 표현하는 데 사용되는 다른 개념이지만, 각각의 장단점과 적합한 상황에 따라 사용됩니다. 상속은 계층 구조와 확장성이 중요한 경우에 유용하며, 합성은 객체 간의 협력과 느슨한 결합도가 필요한 경우에 적합합니다.

상속과 합성은 객체 지향 프로그래밍에서 코드 재사용과 구조 설계에 사용되는 두 가지 주요한 개념입니다.

상속(Inheritance):

차이점:
상속은 클래스 간의 계층적인 관계를 구성하는 개념입니다. 부모 클래스와 자식 클래스 간에 "is-a" 관계가 성립합니다. 자식 클래스는 부모 클래스의 특성과 동작을 상속받아 사용할 수 있습니다.
자식 클래스는 부모 클래스의 멤버(필드, 메서드)를 상속받고 확장, 수정 또는 재정의할 수 있습니다.
공통점:
코드 재사용과 확장성을 제공합니다.
부모 클래스의 멤버를 자식 클래스에서 사용할 수 있습니다.




```
예시:
class Animal {
    protected String name;

    public Animal(String name) {
        this.name = name;
    }

    public void eat() {
        System.out.println(name + " is eating.");
    }
}

class Dog extends Animal {
    public Dog(String name) {
        super(name);
    }

    public void bark() {
        System.out.println(name + " is barking.");
    }
}

public class Main {
    public static void main(String[] args) {
        Dog dog = new Dog("Bobby");
        dog.eat();  // 상속받은 eat() 메서드 호출
        dog.bark(); // 자식 클래스의 bark() 메서드 호출
    }
}

```
합성(Composition):

차이점:
합성은 객체 간의 구성 관계를 나타내는 개념입니다. 하나의 클래스가 다른 클래스의 객체를 포함하고 있는 경우에 사용됩니다. 클래스 간에 "has-a" 관계가 성립합니다.
포함된 객체는 독립적으로 존재할 수 있으며, 포함하는 객체의 수명과 동작에 의존합니다.
공통점:
코드 재사용과 모듈화를 제공합니다.
포함된 객체의 메서드를 호출하여 기능을 사용할 수 있습니다.


```
예시:
class Engine {
    public void start() {
        System.out.println("Engine is starting.");
    }
}

class Car {
    private Engine engine;

    public Car() {
        this.engine = new Engine();
    }

    public void startCar() {
        engine.start();
    }
}

public class Main {
    public static void main(String[] args) {
        Car car = new Car();
        car.startCar(); // 합성된 Engine 객체의 start() 메서드 호출
    }
}

```

위의 예시에서 Car 클래스는 Engine 객체를 포함하고 있으며, Car 클래스의 startCar() 메서드를 호출하면 합성된 Engine 객체의 start() 메서드가 실행됩니다. Car 클래스는 Engine 클래스의 기능을 활용하여 자동차를 시작합니다.




##**16. jvm에 대한 설명과 jvm의 역할에 대해서**

JVM(Java Virtual Machine)은 자바 가상 머신의 약어로, 자바 프로그램을 실행하기 위한 가상 환경을 제공하는 소프트웨어입니다. JVM은 자바 언어로 작성된 프로그램을 특정 운영체제에 종속되지 않고 실행할 수 있도록 합니다.

JVM의 주요 역할은 다음과 같습니다:

자바 프로그램 실행: JVM은 자바 바이트 코드(.class 파일)를 기계어로 변환하고 실행합니다. 자바는 컴파일된 바이트 코드를 JVM이 이해할 수 있는 형태로 변환한 후, JVM에서 실행됩니다.

메모리 관리: JVM은 자동 메모리 관리를 수행합니다. 가비지 컬렉션(Garbage Collection)이라는 기능을 통해 더 이상 사용되지 않는 객체들을 자동으로 제거하여 메모리를 관리합니다.

자원 관리: JVM은 프로그램이 사용하는 다양한 자원(파일, 네트워크 연결 등)을 관리합니다. 프로그램이 자원을 올바르게 사용하고 해제할 수 있도록 지원합니다.

예외 처리: JVM은 자바 프로그램에서 발생하는 예외를 처리합니다. 예외 처리 기능을 통해 프로그램이 예외 상황에 대응하고 복구할 수 있습니다.

스레드 관리: JVM은 멀티스레드를 지원하여 여러 개의 스레드가 동시에 실행될 수 있도록 합니다. 스레드 스케줄링, 동기화 등 스레드 관련 기능을 제공합니다.

JVM은 플랫폼 독립성을 제공하고, 자바 프로그램의 이식성과 성능을 향상시키는 역할을 합니다. 운영체제나 하드웨어에 종속되지 않고 동일한 자바 프로그램을 다양한 환경에서 실행할 수 있으며, JVM은 각 환경에 맞는 최적화와 성능 향상을 위한 기능을 제공합니다.

##**17. 다형성에 대해서**

다형성(Polymorphism)은 객체 지향 프로그래밍의 중요한 개념으로, 하나의 인터페이스나 클래스를 다양한 방식으로 동작하도록 할 수 있는 능력을 말합니다. 다형성을 통해 코드의 유연성과 재사용성을 높일 수 있습니다.

다형성의 주요한 특징은 다음과 같습니다:

오버로딩(Overloading): 동일한 메서드 이름을 가지고 매개변수의 타입 또는 개수를 다르게하여 여러 버전의 메서드를 정의할 수 있습니다.

오버라이딩(Overriding): 부모 클래스의 메서드를 자식 클래스에서 재정의하여 자식 클래스의 동작을 변경할 수 있습니다. 상속 관계에서 발생하는 개념입니다.

인터페이스 구현(Interface Implementation): 인터페이스를 구현하는 클래스들은 같은 인터페이스 타입으로 다룰 수 있습니다. 이를 통해 다양한 구현체를 동일한 인터페이스로 다룰 수 있습니다.

다형적 변수(Polymorphic Variables): 부모 클래스의 변수로 자식 클래스의 객체를 참조할 수 있습니다. 이를 통해 여러 개의 서로 다른 클래스 객체를 하나의 변수로 관리할 수 있습니다.

다형성의 장점은 다음과 같습니다:

유연성과 확장성: 새로운 클래스를 추가하거나 기존 클래스를 수정하지 않고도 코드를 확장하고 변형할 수 있습니다.
코드의 간결성: 동일한 동작을 수행하는 여러 클래스를 단일한 인터페이스로 처리할 수 있습니다.
재사용성: 공통된 기능을 가진 클래스들을 부모 클래스로 추상화하여 재사용할 수 있습니다.


```
예시:
interface Animal {
    void sound();
}

class Dog implements Animal {
    public void sound() {
        System.out.println("멍멍!");
    }
}

class Cat implements Animal {
    public void sound() {
        System.out.println("야옹~");
    }
}

public class Main {
    public static void main(String[] args) {
        Animal dog = new Dog();
        Animal cat = new Cat();

        dog.sound(); // 멍멍!
        cat.sound(); // 야옹~
    }
}

```
위의 예시에서 Animal 인터페이스를 구현한 Dog 클래스와 Cat 클래스는 동일한 인터페이스를 사용하면서 각각 다른 소리를 내는 다형적인 객체입니다. Main 클래스에서 Animal 타입으로 dog와 cat 객체를 다루고 있으며, 호출하는 sound() 메서드는 각 객체의 실제 타입에 맞게 실행됩니다. 이를 통해 동일한 인터페이스를 가진 다양한 객체를 일관적으로 다룰 수 있습니다.


##**18. 스프링 mvc에 대해 설명**

스프링 MVC(Spring MVC)는 스프링 프레임워크(Spring Framework)에서 제공하는 웹 애플리케이션 개발을 위한 모듈입니다. MVC는 Model-View-Controller의 약자로, 애플리케이션의 구조를 세 가지 역할로 분리하여 개발하는 디자인 패턴입니다. 스프링 MVC는 이 패턴을 기반으로 웹 애플리케이션을 구축하고 관리하는 데 사용됩니다.

스프링 MVC의 주요 컴포넌트는 다음과 같습니다:

모델(Model): 애플리케이션의 비즈니스 로직을 처리하는 부분으로, 데이터의 상태와 동작을 정의합니다. 주로 Java 객체로 표현되며, 비즈니스 로직의 처리를 담당합니다.

뷰(View): 모델의 데이터를 사용자에게 보여주는 역할을 담당합니다. 주로 HTML, JSP, Thymeleaf 등의 템플릿 엔진을 사용하여 데이터를 렌더링합니다.

컨트롤러(Controller): 사용자의 요청을 처리하고, 적절한 모델을 호출하여 데이터를 처리한 후 적절한 뷰로 전달하는 역할을 합니다. 요청을 받아 파라미터 검증, 비즈니스 로직 호출, 데이터 전달 등의 작업을 수행합니다.

스프링 MVC의 동작 순서는 다음과 같습니다:

사용자가 웹 브라우저를 통해 요청을 보냅니다.
요청은 DispatcherServlet이라는 스프링 MVC의 핵심 컴포넌트에 도달합니다.
DispatcherServlet은 Handler Mapping을 통해 어떤 컨트롤러가 요청을 처리할지 결정합니다.
선택된 컨트롤러는 요청을 처리하고, 필요한 비즈니스 로직을 수행합니다.
컨트롤러는 모델을 생성하고, 필요한 데이터를 모델에 저장합니다.
컨트롤러는 뷰를 선택하여 모델과 함께 전달합니다.
뷰는 모델의 데이터를 사용하여 사용자에게 결과를 렌더링합니다.
최종적으로 DispatcherServlet은 렌더링된 결과를 사용자에게 응답합니다.

```
@Controller
@RequestMapping("/users")
public class UserController {
    
    @Autowired
    private UserService userService;
    
    @GetMapping
    public String getUsers(Model model) {
        List<User> users = userService.getAllUsers();
        model.addAttribute("users", users);
        return "user-list";
    }
    
    @GetMapping("/{id}")
    public String getUser(@PathVariable("id") Long id, Model model) {
        User user = userService.getUserById(id);
        model.addAttribute("user", user);
        return "user-detail";
    }
    
    @PostMapping
    public String createUser(User user) {
        userService.createUser(user);
        return "redirect:/users";
    }
    
    // 기타 컨트롤러 메서드들...
}

```

위의 예시에서 UserController는 사용자 관리와 관련된 요청을 처리하는 컨트롤러입니다. @GetMapping과 @PostMapping 애노테이션을 사용하여 GET 및 POST 요청을 처리하고, Model 객체를 통해 데이터를 뷰에 전달합니다. 뷰는 Thymeleaf 같은 템플릿 엔진을 사용하여 HTML 페이지를 렌더링합니다. @RequestMapping 애노테이션을 사용하여 /users와 관련된 URL을 매핑합니다.

이를 통해 스프링 MVC는 웹 애플리케이션의 요청을 적절한 컨트롤러로 라우팅하고, 비즈니스 로직을 수행하며, 결과를 뷰로 전달하여 사용자에게 응답하는 구조를 갖추고 있습니다.

##**19. 스프링 aop에 대해서 설명**

스프링 AOP(Aspect-Oriented Programming)는 애플리케이션의 핵심 로직과 공통 기능을 분리하여 모듈화하는 프로그래밍 기법입니다. AOP는 핵심 로직과 공통 기능의 분리를 통해 코드의 재사용성, 유지보수성, 확장성을 향상시킬 수 있습니다.

AOP의 핵심 개념은 '관점(Aspect)'입니다. 관점은 애플리케이션에서 공통으로 발생하는 부가 기능을 말하며, 예를 들어 로깅, 트랜잭션 관리, 보안 등이 될 수 있습니다. 이러한 관점들은 핵심 로직에 영향을 미치지 않으면서 코드 중복을 최소화하고 관심사를 분리하여 모듈화합니다.

스프링 AOP는 프록시 패턴과 다이내믹 프록시를 활용하여 관점 지향 프로그래밍을 구현합니다. 핵심 로직을 수행하는 객체를 대신하여 프록시 객체를 사용하여 관점을 적용합니다. 스프링은 프록시 객체를 자동으로 생성하고 관점을 적용하기 위해 AOP Proxy와 Pointcut, Advice 등의 개념을 제공합니다.



```
@Aspect
@Component
public class LoggingAspect {

    @Before("execution(* com.example.myapp.service.*.*(..))")
    public void beforeMethodExecution(JoinPoint joinPoint) {
        String methodName = joinPoint.getSignature().getName();
        System.out.println("Before executing method: " + methodName);
    }

    @After("execution(* com.example.myapp.service.*.*(..))")
    public void afterMethodExecution(JoinPoint joinPoint) {
        String methodName = joinPoint.getSignature().getName();
        System.out.println("After executing method: " + methodName);
    }
}

@Service
public class UserService {

    public void addUser(String username) {
        System.out.println("Adding user: " + username);
    }
}

```
위의 예시에서 LoggingAspect는 로깅 관점을 정의한 클래스입니다. @Aspect 애노테이션을 사용하여 관점 클래스임을 선언하고, @Before와 @After 애노테이션을 사용하여 핵심 로직의 실행 이전과 이후에 로깅을 수행하는 어드바이스를 정의합니다.

UserService는 핵심 로직을 담당하는 서비스 클래스입니다. addUser 메서드는 사용자를 추가하는 로직을 수행하며, 해당 메서드를 호출할 때 관점인 LoggingAspect의 어드바이스가 적용됩니다. 이를 통해 메서드 실행 이전과 이후에 로깅이 출력됩니다.

스프링 AOP는 이 외에도 다양한 기능을 제공하며, 다른 어드바이스 유형(around, after-returning, after-throwing)과 포인트컷 표현식 등을 사용하여 다양한 관점을 적용할 수 있습니다.


##**20. spring 에서 di에 대해서 설명**

스프링의 DI(Dependency Injection)는 객체 간의 의존성을 외부에서 설정하여 객체가 직접 의존하는 객체를 생성하거나 관리하는 것이 아니라, 외부에서 의존 객체를 주입받는 방식을 말합니다. 이를 통해 객체 간의 결합도를 낮추고 유연하고 재사용 가능한 애플리케이션을 개발할 수 있습니다.

DI의 핵심 아이디어는 '제어의 역전(IoC, Inversion of Control)'입니다. 기존의 객체 생성이나 의존 객체의 관리는 개발자가 직접 수행했지만, DI에서는 객체의 생성과 관리를 스프링 컨테이너가 대신 수행합니다. 개발자는 필요한 의존 객체를 스프링 컨테이너에게 요청하고, 스프링 컨테이너가 의존 객체를 생성하고 주입해줍니다.


```
public class UserService {
    private UserRepository userRepository;

    // 의존 객체를 주입받는 생성자
    public UserService(UserRepository userRepository) {
        this.userRepository = userRepository;
    }

    public void addUser(User user) {
        userRepository.save(user);
    }
}

public class UserRepository {
    public void save(User user) {
        // 사용자 정보를 저장하는 로직
    }
}

@Configuration
public class AppConfig {

    @Bean
    public UserRepository userRepository() {
        return new UserRepository();
    }

    @Bean
    public UserService userService(UserRepository userRepository) {
        return new UserService(userRepository);
    }
}

public class Main {
    public static void main(String[] args) {
        ApplicationContext context = new AnnotationConfigApplicationContext(AppConfig.class);
        UserService userService = context.getBean(UserService.class);

        User user = new User("John Doe");
        userService.addUser(user);
    }
}

```
위의 예시에서 UserService는 UserRepository에 의존하고 있습니다. DI를 적용하기 위해 UserService의 생성자를 통해 UserRepository를 주입받습니다. 이를 통해 UserService는 UserRepository의 메서드를 호출할 수 있습니다.

AppConfig 클래스는 스프링의 설정 클래스로, @Configuration 애노테이션을 사용하여 설정 클래스임을 선언합니다. @Bean 애노테이션을 사용하여 스프링 컨테이너에 빈 객체를 등록합니다. userRepository() 메서드는 UserRepository 객체를 생성하여 반환하고, userService() 메서드는 UserRepository를 주입받는 UserService 객체를 생성하여 반환합니다.

Main 클래스에서는 스프링 컨테이너를 생성하고 UserService 빈 객체를 가져와서 사용합니다. 스프링 컨테이너는 AppConfig 클래스를 참조하여 의존 객체를 생성하고 주입해줍니다. 이를 통해 의존성 주입이 자동으로 이루어지게 됩니다.

DI를 사용함으로써 객체 간의 결합도가 낮아지고 유연한 애플리케이션 개발이 가능해집니다. 또한 테스트 용이성도 향상되며, 객체의 재사용성과 확장성도 증가합니다.


##**21. mvc1과 mvc2패턴의 차이점(잘못된 질문 mvc의 모델1, 모델2가 있음)수정할것**

MVC1과 MVC2는 웹 애플리케이션의 아키텍처 패턴으로, 각각 모델-뷰-컨트롤러(Model-View-Controller) 패턴의 변형입니다. 다음은 MVC1과 MVC2 패턴의 차이점과 각각에 맞는 예시를 설명합니다.

MVC1 패턴:
MVC1은 전통적인 웹 애플리케이션 개발 패턴으로, JSP(JavaServer Pages)와 서블릿(Servlet)을 주로 사용합니다. 주로 JSP가 뷰와 컨트롤러 역할을 모두 담당하고, 서블릿은 모델 역할을 담당합니다. 이 패턴에서는 JSP 내부에 비즈니스 로직과 데이터 처리 로직이 혼재하며, 코드의 재사용성과 유지보수가 어렵습니다.



```
// JSP 파일 (View + Controller 역할)
<%@ page language="java" contentType="text/html; charset=UTF-8" pageEncoding="UTF-8"%>
<%
    String name = request.getParameter("name");
    String greeting = "Hello, " + name + "!";
    out.println(greeting);
%>

// 서블릿 (Model 역할)
@WebServlet("/GreetingServlet")
public class GreetingServlet extends HttpServlet {
    protected void doGet(HttpServletRequest request, HttpServletResponse response) throws ServletException, IOException {
        String name = request.getParameter("name");
        String greeting = "Hello, " + name + "!";
        request.setAttribute("greeting", greeting);
        RequestDispatcher dispatcher = request.getRequestDispatcher("/greeting.jsp");
        dispatcher.forward(request, response);
    }
}

```

MVC2 패턴:
MVC2는 전통적인 MVC 패턴을 웹 애플리케이션에 적용한 것으로, JSP는 뷰 역할을 담당하고, 서블릿은 컨트롤러 역할을 담당합니다. 비즈니스 로직과 데이터 처리 로직은 모델이 담당하며, JSP와 서블릿 간의 역할 분리가 명확해지고, 코드의 재사용성과 유지보수가 용이해집니다. 이를 위해 주로 프레임워크나 라이브러리인 스프링 MVC, Struts 등을 사용합니다.



```
// JSP 파일 (View 역할)
<%@ page language="java" contentType="text/html; charset=UTF-8" pageEncoding="UTF-8"%>
<!DOCTYPE html>
<html>
<head>
    <title>Greeting</title>
</head>
<body>
    <h1>${greeting}</h1>
</body>
</html>

// 서블릿 (Controller 역할)
@WebServlet("/GreetingServlet")
public class GreetingServlet extends HttpServlet {
    protected void doGet(HttpServletRequest request, HttpServletResponse response) throws ServletException, IOException {
        String name = request.getParameter("name");
        String greeting = "Hello, " + name + "!";
        request.setAttribute("greeting", greeting);
        RequestDispatcher dispatcher = request.getRequestDispatcher("/greeting.jsp");
        dispatcher.forward(request, response);
    }
}

```


위 예시에서는 MVC1 패턴에서의 JSP 파일이 뷰 역할을 담당하며, 서블릿이 컨트롤러 역할을 담당합니다. JSP 파일 내부에는 뷰를 구성하는 HTML 코드와 동적으로 생성된 데이터를 출력하는 코드만 포함되어 있습니다. 비즈니스 로직은 서블릿이 처리하며, 결과 데이터를 JSP로 전달하여 화면에 표시합니다. 이렇게 역할을 분리하여 코드의 가독성과 유지보수성을 향상시킬 수 있습니다.
